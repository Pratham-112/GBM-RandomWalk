```python
import numpy as np
import matplotlib.pyplot as plt
# Without drift  price just fluctuates randomly around the starting point.
n_steps = 1000
n_simulations = 5
start_price = 100
# plt.figure(figsize=(10,6))
for i in range(n_simulations):
    steps = np.random.choice([-1, 1], size=n_steps)  # 1 steps
    path = np.cumsum(steps) + start_price
    plt.plot(path, label=f"Simulation {i+1}")
plt.title("Random Walk Simulations Without Drift")
plt.xlabel("Steps")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# With drift
n_steps = 1000
n_simulations = 5
start_price = 100
up_probability=0.55
up_drift=0.1
for i in range(n_simulations):
    steps = np.random.choice([-1, 1], size=n_steps,p=[1-up_probability,up_probability])
    drift_steps=steps+up_drift
    path = np.cumsum(drift_steps) + start_price
    plt.plot(path, label=f"Simulation {i+1}")
plt.title("Random Walk Simulations With Drift")
plt.xlabel("Steps")
plt.ylabel("Price")
plt.legend()
plt.show()
```
